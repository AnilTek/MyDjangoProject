```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
mainData = pd.read_csv('loan_approval_dataset.csv') ## Kullanilan csv dosyasini implement etme
df = mainData.copy() # ve bir kopyasini olusturma
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns = ['loan_id', 'no_of_dependents', 'education', 'self_employed',
       'income_annum', 'loan_amount', 'loan_term', 'cibil_score',
       'residential_assets_value', 'commercial_assets_value',
       'luxury_assets_value', 'bank_asset_value', 'loan_status']
```

```python
adjusted_df=pd.get_dummies(df,columns=["education",'self_employed','loan_status']).drop(['education_ Not Graduate','self_employed_ No','loan_status_ Rejected'],axis=1) 

# get dummies ile string olan degerleri categorical olarak farkli sutunlara bolup drop first ile de teke indiriyorum
```

```python
adjusted_df.head()
```

```python
adjusted_df.columns =['loan_id', 'no_of_dependents',
       'income_annum', 'loan_amount', 'loan_term', 'cibil_score',
       'residential_assets_value', 'commercial_assets_value',
       'luxury_assets_value', 'bank_asset_value','graduate','employed','result'] ## sutun isimlerini tekrardan ayarliyorum cunku sutun isimlerinden 
## olmamasi gereken bosluklar kullanilmis 
```

```python
adjusted_df.head()
```

```python
adjusted_df['graduate'] = adjusted_df['graduate'].astype(int)
adjusted_df['employed'] = adjusted_df['employed'].astype(int)
adjusted_df['result'] = adjusted_df['result'].astype(int)
```

```python
adjusted_df.head()
```

```python
adjusted_df.corr() ## corr yardimi ile sutunlarin birbirileri ile yaptigi positif ve negatif korelasyonlari gorebiliyoruz

## !! NOT CIBIL_SCORE UN RESULT A ETKISI BUYUK (Cibil puani ise kredi puani )
```

```python
from sklearn.model_selection import train_test_split
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X= adjusted_df.drop(['result','loan_id'],axis=1)
y= adjusted_df['result']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)



X_train, X_test, y_train, y_test = train_test_split(X_resampled,y_resampled , test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
lr_predict = lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,lr_predict))
```

```python
print(confusion_matrix(y_test,lr_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100)
```

```python
rf_model.fit(X_train,y_train)
```

```python
rf_predict = rf_model.predict(X_test)
```

```python
print(classification_report(y_test,rf_predict))
print(confusion_matrix(y_test,rf_predict))
```
